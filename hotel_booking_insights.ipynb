data analysis and visualizations

```python
canceled_per=df['is_canceled'].value_counts(normalize=True)
print(canceled_per)

plt.figure(figsize=(5,4))
plt.title("canceled reservations")
plt.bar(['not_canceled','canceled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
df['hotel'] = df['hotel'].astype("string")
```

```python
df.info()
```

```python
import numpy as np
```

```python
grouped_data = df.groupby(['hotel', 'is_canceled']).size().unstack(fill_value=0)

# Set the figure size
plt.figure(figsize=(8, 4))

# Get the positions and width for the bars
positions = np.arange(len(grouped_data))
width = 0.4

# Plot the data
plt.bar(positions - width/2, grouped_data[0], width=width, label='Not Canceled', color='blue')
plt.bar(positions + width/2, grouped_data[1], width=width, label='Canceled', color='lightblue')

# Add title and labels
plt.title('Reservation Status in Different Hotels', size=20)
plt.xlabel('Hotel', size=12)
plt.ylabel('Number of Reservations', size=12)
plt.xticks(positions, grouped_data.index)
plt.legend(bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

```python
resort_hotel=df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby( 'reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean ()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label= 'Resort hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label= 'City hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df[ 'month'] = df[ 'reservation_status_date'].dt .month
grouped_data = df.groupby(['month', 'is_canceled']).size().unstack(fill_value=0)

# Set the figure size
plt.figure(figsize=(16, 8))

# Get the positions and width for the bars
positions = np.arange(len(grouped_data))
width = 0.4

# Plot the data
plt.bar(positions - width/2, grouped_data[0], width=width, label='Not Canceled', color='blue')
plt.bar(positions + width/2, grouped_data[1], width=width, label='Canceled', color='orange')

# Add title and labels
plt.title('Reservation Status per Month', size=20)
plt.xlabel('Month', size=12)
plt.ylabel('Number of Reservations', size=12)
plt.xticks(positions, grouped_data.index)
plt.legend(bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

```python
grouped_data = df[df['is_canceled'] == 1].groupby('month')['adr'].sum().reset_index()

# Set the figure size
plt.figure(figsize=(15, 8))

colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightseagreen', 'lightpink', 'lightblue', 
          'lightgray', 'lightgoldenrodyellow', 'lightcyan', 'lightyellow', 'lightsteelblue']

# Plot the data using bar plot
plt.bar(grouped_data['month'], grouped_data['adr'], color=colors)

# Add title and labels
plt.title('ADR per Month', fontsize=30)
plt.xlabel('Month', fontsize=15)
plt.ylabel('ADR', fontsize=15)

# Show the plot
plt.show()
```

```python
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data[ 'country']. value_counts () [:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with reservation canceled' )
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt. show()
```
